Fazer um programa que converte a temperatura de celsius para fahrenheit.

#Entrada: graus celsius

#Processamento: converter para fahrenheit

#Saida: mostrar a temperatura em fahrenheit


```python
C = float(input("Digite a temperatura em graus celsius: "))
F = (C * 9/5) + 32
print(f"A temperatura em fahrenheit é: {F}")
```

```python
# Operadores aritiméticos

# Soma de dois números

a = int(input("Digite o primeiro número: "))
b = int(input("Digite o segundo número: "))
soma = a + b
subtraçao = a - b
multiplicaçao = a * b
divisao = a // b
resto = a % b # resto da divisão
print(f"A soma de {a} com {b} é igual a: {soma} ")
print(f"A diferença de {a} e {b} é igual a: {subtraçao} ")
print(f"A multiplicação de {a} com {b} é igual a: {multiplicaçao} ")
print(f"A divisão de {a} dividido por {b} é igual a: {divisao} ")
print(f"A divisão de {a} dividido {b} o resto é igual a: {resto} ")
```

```python
#operadores relacionais
num1 = int(input("Digite um número: "))
num2 = int(input("Digite um número: "))
igual = num1 == num2
maior = num1 > num2
menor = num1 < num2
mi = num1 <= num2
maiori = num1 >= num2
diferente = num1 != num2
print(igual)
print(maior)
print(menor)
print(mi)
print(maiori)
print(diferente)
```

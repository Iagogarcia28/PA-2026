Escrever um algoritmo e o programa que recebe 04 notas do aluno e calcula a média depois mostrar a média

```python
n1 = float(input("digite a primeira nota: "))
n2 = float(input("digite a segunda nota: "))
n3 = float(input("digite a terceira nota: "))
n4 = float(input("digite a quarta nota: "))
md = (n1 + n2 + n3 + n4) / 4
print(f"a media do aluno é: {md}")
```
